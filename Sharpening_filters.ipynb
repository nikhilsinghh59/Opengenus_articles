```
# Importing libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
# Read input image
img = cv2.imread('img2.jpg')
img = cv2.resize(img , (400 , 400))
#print(img.shape)


# Display the input image
cv2_imshow(img)
```

```
# lets make a kernel which we will slide over our image(convolution)

kernel = np.array([[-1 , -1 , -1] , 
                   [-1 , 9 , -1] , 
                   [-1 , -1 , -1]])

print(kernel)
```

```
# Now we will perform convolution using open cv for shapening ur image
sharp_img = cv2.filter2D(img , -1 , kernel = kernel)

cv2_imshow(sharp_img)
```

**Using PIL library**

PIL is a python image processing library.

ImageFilter module in PIL has some pre defined set of kernels for each filter and uses those filters for applying convolution operation.

```
# Importing libraries
from PIL import Image
from PIL import ImageFilter
```

```
# Open image using Image class
img = Image.open('img3.png')
img
```

```
# Applying sharpening filter to our image
sharp1 = img.filter(ImageFilter.SHARPEN)
sharp1
```

```
sharp2 = sharp1.filter(ImageFilter.SHARPEN)
sharp2
```
